```python
def clean_data(file):
    df = pd.read_csv(file)

    df = df.fillna('0',inplace=True)
    df["Deaths"].fillna('0' ,inplace=True)
    df["Confirmed"].fillna('0',inplace=True)
    df["Mortality Rate"] = (
        df["Deaths"] / df["Confirmed"]
    ) * 100

    return df
```

```python
df=clean_data("covid_19_clean_complete.csv")
df.head()
```

```python
country_case=df.groupby("Country/Region")["Confirmed"].sum()
print(country_case)
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = df["Date"].dt.month
df.groupby("Month")["Confirmed"].sum()
```

```python
agg=df.groupby("WHO Region")["Deaths"].agg(["sum","mean","max"])
print(agg)
```

```python
df["7-Day Average"] = df["Confirmed"].rolling(7).mean()
```

```python
print(df["7-Day Average"])
```

```python
df["Cumulative Cases"] = df["Confirmed"].cumsum()

print(df[["Date", "Confirmed", "Cumulative Cases"]].tail().head())
```

```python
df["Daily Change"] = df["Confirmed"].diff()

print(df[["Date", "Daily Change"]].head())
```

```python
df.to_csv('output/result.csv', index=False)
print("file saved")
```

```python
df['Date'] = pd.to_datetime(df['Date'])
China= df[df['Country/Region'] == 'China']
plt.figure(figsize=(12,6))
plt.bar(China['Date'], China['Confirmed'])
plt.title('COVID-19 Confirmed Cases in China')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.grid(True)
plt.show()
```

```python
countries = ['India', 'United States', 'Brazil', 'United Kingdom']

plt.figure(figsize=(14,7))

for country in countries:
    data = df[df['Country/Region'] == country]
    plt.semilogy( data['Date'], data['Deaths'], label=country)
    plt.plot(data['Date'], data['Deaths'], label=country)

plt.title('COVID-19 Death Trends')
plt.xlabel('Date')
plt.ylabel('Deaths')
plt.legend()
plt.grid(True)
plt.show()
```

```python

from pptx import Presentation
from pptx.util import Inches
import matplotlib.pyplot as plt
import pandas as pd


# Create PPT object
ppt = Presentation()

# ---------------- Slide 1: Title ----------------
slide = ppt.slides.add_slide(ppt.slide_layouts[0])
title = slide.shapes.title
subtitle = slide.placeholders[1]

title.text = "COVID-19 Data Analysis"
subtitle.text = "Python, Pandas, Matplotlib Project"

# ----------------Function to save plot----------------
def save_plot(filename):
    plt.savefig(filename)
    plt.close()

# ---------------- Plot 1: Confirmed Cases Trend ----------------
plt.figure(figsize=(6,4))
plt.plot(df["Date"], df["Confirmed"])
plt.title("COVID-19 Confirmed Cases Trend")
plt.xticks(rotation=45)
save_plot("plot1.png")

slide = ppt.slides.add_slide(ppt.slide_layouts[5])
slide.shapes.title.text = "Confirmed Cases Trend"
slide.shapes.add_picture("plot1.png", Inches(1), Inches(1.5), width=Inches(8))

# ---------------- Plot 2: Deaths Trend ----------------
plt.figure(figsize=(6,4))
plt.plot(df["Date"], df["Deaths"])
plt.title("COVID-19 Deaths Trend")
plt.xticks(rotation=45)
save_plot("plot2.png")

slide = ppt.slides.add_slide(ppt.slide_layouts[5])
slide.shapes.title.text = "Deaths Trend"
slide.shapes.add_picture("plot2.png", Inches(1), Inches(1.5), width=Inches(8))

# ---------------- Plot 3: Recovery Trend ----------------
plt.figure(figsize=(6,4))
plt.plot(df["Date"], df["Recovered"])
plt.title("COVID-19 Recovery Trend")
plt.xticks(rotation=45)
save_plot("plot3.png")

slide = ppt.slides.add_slide(ppt.slide_layouts[5])
slide.shapes.title.text = "Recovery Trend"
slide.shapes.add_picture("plot3.png", Inches(1), Inches(1.5), width=Inches(8))

# ---------------- Plot 4: Country Comparison ----------------
country_data = df.groupby("Country/Region")["Confirmed"].sum().head(10)

plt.figure(figsize=(6,4))
country_data.plot(kind="bar")
plt.title("Top 10 Countries - Confirmed Cases")
save_plot("plot4.png")

slide = ppt.slides.add_slide(ppt.slide_layouts[5])
slide.shapes.title.text = "Country Comparison"
slide.shapes.add_picture("plot4.png", Inches(1), Inches(1.5), width=Inches(8))

# ---------------- Save PPT ----------------
ppt.save("COVID19_Analysis_Presentation.pptx")

print("PPT Created Successfully!")
```
